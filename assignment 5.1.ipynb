```python
import numpy as np
import pandas as pd
li=[110,105,115,110,130,150,100,105]
```

```python
df = pd.DataFrame(li, columns=['price'])
df
```

# Normalization / min max scaler

```python
from sklearn.preprocessing import MinMaxScaler
mmx = MinMaxScaler() 
df['price_after_normalization'] = mmx.fit_transform(df[['price']])
df
```

# Standardization / standard scaler

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
df['price_after_standardization'] = scaler.fit_transform(df[['price']])
df
```

# Log Transformation

```python
import numpy as np
df['fc_numpy'] = np.log1p(df[['price']]) 
df 
```

#  Robust Scaler

```python
from sklearn.preprocessing import MaxAbsScaler
mx = MaxAbsScaler()
df['max_abs_scaler'] = mx.fit_transform(df[['price']])
df
```

```python
data=pd.DataFrame(df)
data.to_excel("Assignment_5.1.xlsx")
data
```
